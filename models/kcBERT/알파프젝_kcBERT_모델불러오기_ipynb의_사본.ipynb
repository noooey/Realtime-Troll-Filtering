```python
import torch
import numpy as np

def new_softmax(a):
    # 입력 배열에서 logits 추출
    logits = a
    
    # logits에 softmax 함수 적용
    y = torch.nn.functional.softmax(logits, dim=-1)
    return y.tolist()


# 예제 입력 문장
input_text = ""

# 입력 문장을 토큰화하여 텐서로 변환
inputs = tokenizer(input_text, return_tensors="pt")
input_ids = inputs["input_ids"]
attention_mask = inputs["attention_mask"]

# 모델의 출력 크기 확인
outputs = model(input_ids=input_ids, attention_mask=attention_mask)
logits = outputs.last_hidden_state
total = 0
probability = []
logits = np.round(new_softmax(logits), 3).tolist()
for logit in logits:
    probability.append(logit)

print("logits")
print(logits)
print(np.max(logits))
if np.max(logits) <= 0.5:
    emotion = "general"
elif np.max(logits) >= 0.5:
    emotion = "hate"

probability.append(emotion)
print("prob", probability)

# 예측 결과 출력
preds = torch.softmax(torch.tensor(logits), dim=-1)
print(preds)  # tensor([[0.0208, 0.9792]], grad_fn=<SoftmaxBackward>)
print(emotion)
```
